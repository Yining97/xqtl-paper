# SEM Mediation Analysis: Parallel Design (4 Mediators)
## APOE Region SNP -> Gene Expression -> CAA Neuropathology

**Aim:** Test whether the effect of the APOE-region SNP `chr19_44938650_G_C` on cerebral amyloid angiopathy 
(CAA, neocortical score `caa_neo4`) is mediated through gene expression of four genes in a **parallel mediation** model:

1. **APOC4/APOC2** (anterior cingulate cortex expression)
2. **APOC2** (anterior cingulate cortex expression)  
3. **APOC1** (De Jager microglia expression)
4. **APOE** (Mega microglia expression)

All four mediators enter one simultaneous model so we estimate **specific indirect effects** 
(the unique mediation contribution of each gene, controlling for the other three) and the **total indirect effect**.

**Design:** Parallel mediation (Design 2), unidirectional: SNP -> {M1, M2, M3, M4} -> Y

**Dataset:** Set 27, N = 1153 subjects with partially overlapping phenotype availability.


## Directed Acyclic Graph (DAG)

```
                          a1                     b1
                    +--------> [APOC4_APOC2] --------+
                    |     a2                     b2   |
                    +--------> [APOC2]      --------+ |
  [chr19_44938650]--+                               +---> [caa_neo4]
                    |     a3                     b3   |
                    +--------> [APOC1_Mic]  --------+ |
                    |     a4                     b4   |
                    +--------> [APOE_Mic]   ---------+
        |                                              ^
        |                  c' (direct)                 |
        +----------------------------------------------+
```

**Specific indirect effects:**
- ind1 = a1 x b1 (via APOC4/APOC2 AC expression)
- ind2 = a2 x b2 (via APOC2 AC expression)
- ind3 = a3 x b3 (via APOC1 DeJager microglia expression)
- ind4 = a4 x b4 (via APOE Mega microglia expression)

**Total indirect** = ind1 + ind2 + ind3 + ind4  
**Total effect** = c' + total_indirect

**Covariates:** Each mediator equation includes its own covariates (Strategy A: covariates-in-model).
- M1 (APOC4_APOC2_AC): msex_u, age_death_u, pmi_u, ROS_study_u
- M2 (APOC2_AC): msex_u, age_death_u, pmi_u, ROS_study_u
- M3 (APOC1_DeJager_Mic): msex_u, age_death_u, pmi_u
- M4 (APOE_Mega_Mic): msex_u, age_death_u, pmi_u
- Y (caa_neo4): educ, apoe4_dose, apoe2_dose, msex_u, age_death_u

**Residual correlations** M_i ~~ M_j are freely estimated for all mediator pairs.

## 4. Parallel Mediation Model Specification

**Strategy A: Covariates-in-model.** Each mediator equation includes its own covariates. 
The outcome equation includes its own covariates. FIML handles the different missingness patterns.

**Residual correlations:** All 6 pairwise mediator residual covariances (M_i ~~ M_j) are freely estimated,
acknowledging shared variance not explained by the SNP.


````r
# ============================================================
# Build parallel mediation model string
# ============================================================
n_med <- length(MED_COLS)

# Mediator equations: Mi ~ ai * SNP + covariates
med_eqs <- character(n_med)
for (i in 1:n_med) {
    cov_str <- paste(MED_COVS[[paste0("M", i)]], collapse = " + ")
    med_eqs[i] <- paste0(MED_COLS[i], " ~ a", i, " * ", SNP_COL, " + ", cov_str)
}

# Outcome equation: Y ~ b1*M1 + b2*M2 + b3*M3 + b4*M4 + cp*SNP + covariates
y_med_terms <- paste0("b", 1:n_med, " * ", MED_COLS, collapse = " + ")
out_cov_str <- paste(OUT_COVS, collapse = " + ")
y_eq <- paste0(OUT_COL, " ~ ", y_med_terms, " + cp * ", SNP_COL, " + ", out_cov_str)

# Specific indirect effects
ind_defs <- paste0("ind", 1:n_med, " := a", 1:n_med, " * b", 1:n_med)

# Total indirect and total
total_ind <- paste0("total_indirect := ", paste0("ind", 1:n_med, collapse = " + "))
total_def <- "total := cp + total_indirect"

# Proportion mediated
prop_med_def <- "prop_med := total_indirect / total"

# Pairwise contrasts
contrasts <- c()
for (i in 1:(n_med-1)) {
    for (j in (i+1):n_med) {
        contrasts <- c(contrasts, paste0("diff_", i, "_", j, " := ind", i, " - ind", j))
    }
}

# Residual correlations between mediators (explicitly)
cov_terms <- c()
for (i in 1:(n_med-1)) {
    for (j in (i+1):n_med) {
        cov_terms <- c(cov_terms, paste0(MED_COLS[i], " ~~ ", MED_COLS[j]))
    }
}

model_str <- paste(c(med_eqs, y_eq, ind_defs, total_ind, total_def, prop_med_def, contrasts, cov_terms), 
                   collapse = "\n")

cat("=== Parallel Mediation Model ===\n\n")
cat(model_str, "\n")
````

## 5. Method 1: FIML SEM (Full Information Maximum Likelihood)

FIML uses **all N subjects** who have the exposure observed, even if they are missing one or more 
mediators or the outcome. This maximizes statistical power by leveraging partially overlapping samples.
`fixed.x=FALSE` ensures the SNP is modeled as a random variable, allowing subjects with only the SNP 
observed to contribute information to the model.


````r
# ============================================================
# FIML SEM fit
# ============================================================
cat("Fitting FIML SEM with N =", N_total, "...\n\n")

fit_fiml <- tryCatch(
    sem(model_str, data = dat_sub, missing = "fiml", fixed.x = FALSE, estimator = "ML"),
    error = function(e) { cat("FIML Error:", conditionMessage(e), "\n"); NULL }
)

if (!is.null(fit_fiml) && lavInspect(fit_fiml, "converged")) {
    cat("FIML converged successfully.\n")
    cat("N used:", lavInspect(fit_fiml, "nobs"), "\n\n")
    
    # Summary
    pe <- parameterEstimates(fit_fiml, ci = TRUE)
    
    # Extract key labeled parameters
    key_labels <- c(paste0("a", 1:n_med), paste0("b", 1:n_med), "cp",
                    paste0("ind", 1:n_med), "total_indirect", "total", "prop_med",
                    paste0("diff_", apply(combn(n_med, 2), 2, paste, collapse="_"), ""))
    # Fix contrast labels
    contrast_labels <- c()
    for (i in 1:(n_med-1)) {
        for (j in (i+1):n_med) {
            contrast_labels <- c(contrast_labels, paste0("diff_", i, "_", j))
        }
    }
    key_labels <- c(paste0("a", 1:n_med), paste0("b", 1:n_med), "cp",
                    paste0("ind", 1:n_med), "total_indirect", "total", "prop_med",
                    contrast_labels)
    
    key_pe <- pe[pe$label %in% key_labels, c("label", "est", "se", "ci.lower", "ci.upper", "z", "pvalue")]
    key_pe$N <- N_total
    key_pe <- key_pe[match(key_labels[key_labels %in% key_pe$label], key_pe$label), ]
    
    print(key_pe, digits=4, row.names=FALSE)
    
    # Save
    write.csv(key_pe, file.path(fiml_dir, "fiml_all_paths.csv"), row.names = FALSE)
    cat("\nSaved:", file.path(fiml_dir, "fiml_all_paths.csv"), "\n")
    
    # Also save fit measures
    fm <- fitMeasures(fit_fiml, c("chisq", "df", "pvalue", "cfi", "tli", "rmsea", "srmr", "aic", "bic"))
    fm_df <- data.frame(measure = names(fm), value = as.numeric(fm), N = N_total)
    write.csv(fm_df, file.path(fiml_dir, "fiml_fit_measures.csv"), row.names = FALSE)
} else {
    cat("FIML did not converge!\n")
}
````

````r
# ============================================================
# FIML: Residual correlations between mediators
# ============================================================
if (!is.null(fit_fiml) && lavInspect(fit_fiml, "converged")) {
    rescov <- pe[pe$op == "~~" & pe$lhs %in% MED_COLS & pe$rhs %in% MED_COLS & pe$lhs != pe$rhs,
                 c("lhs", "rhs", "est", "se", "ci.lower", "ci.upper", "z", "pvalue")]
    cat("\n=== Residual Correlations Between Mediators ===\n")
    print(rescov, digits=4, row.names=FALSE)
    write.csv(rescov, file.path(fiml_dir, "fiml_mediator_residual_covariances.csv"), row.names = FALSE)
}
````

### FIML Interpretation

**Path estimates:**
- **a-paths (a1-a4):** Effect of the SNP on each mediator (gene expression), controlling for mediator-specific covariates
- **b-paths (b1-b4):** Effect of each mediator on CAA outcome, controlling for the SNP, other mediators, and outcome covariates
- **c' (direct):** Direct effect of SNP on CAA not through any of the 4 mediators
- **ind1-ind4:** Specific indirect effects -- each represents the unique mediation contribution of that gene
- **total_indirect:** Sum of all specific indirect effects
- **total:** c' + total_indirect (total SNP effect on CAA)
- **prop_med:** Proportion of total effect mediated through all 4 genes combined
- **Contrasts (diff_i_j):** Differences between specific indirect effects (is one mediator stronger than another?)
- **Residual covariances:** Shared variance between mediators beyond what the SNP explains


````r
# ============================================================
# FIML Forest Plot
# ============================================================
if (!is.null(fit_fiml) && lavInspect(fit_fiml, "converged")) {
    # Prepare plot data: main paths only (not contrasts)
    plot_labels <- c(paste0("a", 1:n_med), paste0("b", 1:n_med), "cp",
                     paste0("ind", 1:n_med), "total_indirect", "total", "prop_med")
    plot_pe <- key_pe[key_pe$label %in% plot_labels, ]
    
    # Nice display names
    display_names <- c(
        a1 = "a1: SNP->APOC4/APOC2_AC", a2 = "a2: SNP->APOC2_AC",
        a3 = "a3: SNP->APOC1_Mic", a4 = "a4: SNP->APOE_Mic",
        b1 = "b1: APOC4/APOC2_AC->CAA", b2 = "b2: APOC2_AC->CAA",
        b3 = "b3: APOC1_Mic->CAA", b4 = "b4: APOE_Mic->CAA",
        cp = "c': SNP->CAA (direct)",
        ind1 = "ind1: via APOC4/APOC2_AC", ind2 = "ind2: via APOC2_AC",
        ind3 = "ind3: via APOC1_Mic", ind4 = "ind4: via APOE_Mic",
        total_indirect = "Total indirect", total = "Total effect", prop_med = "Prop. mediated"
    )
    
    plot_pe$display <- display_names[plot_pe$label]
    plot_pe$display <- factor(plot_pe$display, levels = rev(display_names[plot_labels]))
    
    # Effect type for coloring
    plot_pe$effect_type <- ifelse(grepl("^a", plot_pe$label), "a-path (SNP->M)",
                           ifelse(grepl("^b", plot_pe$label), "b-path (M->Y)",
                           ifelse(plot_pe$label == "cp", "Direct (c')",
                           ifelse(grepl("^ind", plot_pe$label), "Specific indirect",
                           ifelse(plot_pe$label == "total_indirect", "Total indirect",
                           ifelse(plot_pe$label == "total", "Total effect", "Proportion"))))))
    
    p_fiml <- ggplot(plot_pe, aes(x = est, y = display, color = effect_type)) +
        geom_vline(xintercept = 0, linetype = "dashed", color = "grey50") +
        geom_pointrange(aes(xmin = ci.lower, xmax = ci.upper), size = 0.5) +
        scale_color_brewer(palette = "Set1") +
        labs(title = "FIML SEM: Parallel Mediation Path Estimates",
             subtitle = paste0("SNP -> {APOC4/APOC2, APOC2, APOC1, APOE} -> caa_neo4 | N = ", N_total, " (FIML)"),
             x = "Estimate (95% CI)", y = "", color = "Effect Type") +
        theme_minimal(base_size = 12) +
        theme(legend.position = "bottom")
    
    ggsave(file.path(fiml_dir, "fiml_forest_plot.png"), p_fiml, width = 11, height = 9, dpi = 150)
    ggsave(file.path(fiml_dir, "fiml_forest_plot.pdf"), p_fiml, width = 11, height = 9)
    print(p_fiml)
    cat("\nFIML forest plot saved.\n")
}
````

### Sensitivity Analysis: FIML with 3 Mediators (Excluding APOC4_APOC2_AC)

Because APOC4_APOC2_AC_exp and APOC2_AC_exp showed large opposing b-path estimates with wide CIs in the 4-mediator model (suggesting collinearity), we re-run the FIML analysis excluding APOC4_APOC2_AC_exp. This tests whether the remaining 3 mediators show clearer indirect effects when freed from collinearity.

````r
# ============================================================
# Sensitivity: FIML SEM with 3 Mediators (excl. APOC4_APOC2_AC)
# ============================================================
sens_dir <- file.path(fiml_dir, "sensitivity_analysis")
dir.create(sens_dir, recursive = TRUE, showWarnings = FALSE)

# 3-mediator setup
MED_COLS_3 <- c("APOC2_AC_exp", "APOC1_DeJager_Mic_exp", "APOE_Mega_Mic_exp")
MED_SHORT_3 <- c("APOC2_AC", "APOC1_Mic", "APOE_Mic")
n_med_3 <- length(MED_COLS_3)

MED_COVS_3 <- list(
    M1 = c("msex_u", "age_death_u", "pmi_u", "ROS_study_u"),   # APOC2_AC
    M2 = c("msex_u", "age_death_u", "pmi_u"),                   # APOC1_Mic
    M3 = c("msex_u", "age_death_u", "pmi_u")                    # APOE_Mic
)

# Build model
med_eqs_3 <- character(n_med_3)
for (i in 1:n_med_3) {
    cov_str <- paste(MED_COVS_3[[paste0("M", i)]], collapse = " + ")
    med_eqs_3[i] <- paste0(MED_COLS_3[i], " ~ a", i, " * ", SNP_COL, " + ", cov_str)
}
y_med_terms_3 <- paste0("b", 1:n_med_3, " * ", MED_COLS_3, collapse = " + ")
y_eq_3 <- paste0(OUT_COL, " ~ ", y_med_terms_3, " + cp * ", SNP_COL, " + ", out_cov_str)

ind_defs_3 <- paste0("ind", 1:n_med_3, " := a", 1:n_med_3, " * b", 1:n_med_3)
total_ind_3 <- paste0("total_indirect := ", paste0("ind", 1:n_med_3, collapse = " + "))
total_def_3 <- "total := cp + total_indirect"
prop_med_3 <- "prop_med := total_indirect / total"

contrasts_3 <- c()
for (i in 1:(n_med_3 - 1)) for (j in (i+1):n_med_3)
    contrasts_3 <- c(contrasts_3, paste0("diff_", i, "_", j, " := ind", i, " - ind", j))

cov_terms_3 <- c()
for (i in 1:(n_med_3 - 1)) for (j in (i+1):n_med_3)
    cov_terms_3 <- c(cov_terms_3, paste0(MED_COLS_3[i], " ~~ ", MED_COLS_3[j]))

model_str_3 <- paste(c(med_eqs_3, y_eq_3, ind_defs_3, total_ind_3, total_def_3, prop_med_3, contrasts_3, cov_terms_3),
                     collapse = "\n")

cat("=== 3-Mediator Parallel Model (excl. APOC4_APOC2_AC) ===\n\n")
cat(model_str_3, "\n")
````

````r
# ============================================================
# Sensitivity: Fit FIML with 3 mediators
# ============================================================
cat("Fitting 3-mediator FIML SEM with N =", N_total, "...\n\n")

fit_fiml_3 <- tryCatch(
    sem(model_str_3, data = dat_sub, missing = "fiml", fixed.x = FALSE, estimator = "ML"),
    error = function(e) { cat("FIML Error:", conditionMessage(e), "\n"); NULL }
)

if (!is.null(fit_fiml_3) && lavInspect(fit_fiml_3, "converged")) {
    cat("3-mediator FIML converged successfully.\n")
    cat("N used:", lavInspect(fit_fiml_3, "nobs"), "\n\n")

    pe_3 <- parameterEstimates(fit_fiml_3, ci = TRUE)

    contrast_labels_3 <- c()
    for (i in 1:(n_med_3 - 1)) for (j in (i+1):n_med_3)
        contrast_labels_3 <- c(contrast_labels_3, paste0("diff_", i, "_", j))

    key_labels_3 <- c(paste0("a", 1:n_med_3), paste0("b", 1:n_med_3), "cp",
                      paste0("ind", 1:n_med_3), "total_indirect", "total", "prop_med",
                      contrast_labels_3)

    key_pe_3 <- pe_3[pe_3$label %in% key_labels_3, c("label", "est", "se", "ci.lower", "ci.upper", "z", "pvalue")]
    key_pe_3$N <- N_total
    key_pe_3 <- key_pe_3[match(key_labels_3[key_labels_3 %in% key_pe_3$label], key_pe_3$label), ]

    print(key_pe_3, digits = 4, row.names = FALSE)

    write.csv(key_pe_3, file.path(sens_dir, "fiml_3med_all_paths.csv"), row.names = FALSE)
    cat("\nSaved:", file.path(sens_dir, "fiml_3med_all_paths.csv"), "\n")

    # Fit measures
    fm_3 <- fitMeasures(fit_fiml_3, c("chisq", "df", "pvalue", "cfi", "tli", "rmsea", "srmr", "aic", "bic"))
    fm_df_3 <- data.frame(measure = names(fm_3), value = as.numeric(fm_3), N = N_total)
    write.csv(fm_df_3, file.path(sens_dir, "fiml_3med_fit_measures.csv"), row.names = FALSE)

    # Residual covariances
    rescov_3 <- pe_3[pe_3$op == "~~" & pe_3$lhs %in% MED_COLS_3 & pe_3$rhs %in% MED_COLS_3 & pe_3$lhs != pe_3$rhs,
                     c("lhs", "rhs", "est", "se", "ci.lower", "ci.upper", "z", "pvalue")]
    cat("\n=== Residual Correlations Between 3 Mediators ===\n")
    print(rescov_3, digits = 4, row.names = FALSE)
    write.csv(rescov_3, file.path(sens_dir, "fiml_3med_residual_covariances.csv"), row.names = FALSE)
} else {
    cat("3-mediator FIML did not converge!\n")
}
````

## 6. Method 2: Bootstrap (FIML Inside Each Replicate)

We resample the **full dataset** (N = all subjects with exposure observed) and fit the same parallel 
mediation SEM using FIML inside each replicate. This provides nonparametric confidence intervals for 
the specific and total indirect effects.


````r
# ============================================================
# Bootstrap: FIML inside each replicate
# ============================================================
set.seed(12345)
B <- 1000

# All labels to extract
param_labels <- c(paste0("a", 1:n_med), paste0("b", 1:n_med), "cp",
                  paste0("ind", 1:n_med), "total_indirect", "total", "prop_med")

boot_mat <- matrix(NA, B, length(param_labels), dimnames = list(NULL, param_labels))

cat("Running", B, "bootstrap replicates (FIML inside each)...\n")
n_converged <- 0
t0 <- proc.time()[3]

for (i in 1:B) {
    idx <- sample(nrow(dat_sub), replace = TRUE)
    boot_dat <- dat_sub[idx, ]
    
    fit_b <- tryCatch(
        sem(model_str, data = boot_dat, missing = "fiml", fixed.x = FALSE,
            se = "none", test = "none"),
        error = function(e) NULL,
        warning = function(w) {
            tryCatch(
                sem(model_str, data = boot_dat, missing = "fiml", fixed.x = FALSE,
                    se = "none", test = "none"),
                error = function(e) NULL
            )
        }
    )
    
    if (!is.null(fit_b) && lavInspect(fit_b, "converged")) {
        pe_b <- parameterEstimates(fit_b)
        for (lab in param_labels) {
            row_b <- pe_b[pe_b$label == lab, ]
            if (nrow(row_b) == 1) boot_mat[i, lab] <- row_b$est
        }
        n_converged <- n_converged + 1
    }
    
    if (i %% 200 == 0) {
        elapsed <- proc.time()[3] - t0
        cat(sprintf("  Replicate %d/%d (converged: %d, elapsed: %.0fs)\n", i, B, n_converged, elapsed))
    }
}

cat(sprintf("\nBootstrap complete: %d/%d converged (%.1f%%)\n", n_converged, B, 100*n_converged/B))
````

````r
# ============================================================
# Bootstrap: Summarize results
# ============================================================
boot_summary <- data.frame(
    label = param_labels,
    boot_mean = apply(boot_mat, 2, mean, na.rm=TRUE),
    boot_se   = apply(boot_mat, 2, sd, na.rm=TRUE),
    ci_lower  = apply(boot_mat, 2, quantile, 0.025, na.rm=TRUE),
    ci_upper  = apply(boot_mat, 2, quantile, 0.975, na.rm=TRUE),
    stringsAsFactors = FALSE
)

# Two-sided p-value from bootstrap distribution
boot_summary$p_value <- sapply(param_labels, function(lab) {
    vals <- boot_mat[, lab]
    vals <- vals[!is.na(vals)]
    if (length(vals) < 10) return(NA)
    p_pos <- mean(vals > 0)
    2 * min(p_pos, 1 - p_pos)
})

boot_summary$N <- N_total
boot_summary$n_converged <- n_converged
boot_summary$ci_type <- "Percentile"

print(boot_summary, digits=4, row.names=FALSE)

write.csv(boot_summary, file.path(boot_dir, "bootstrap_results.csv"), row.names = FALSE)
cat("\nSaved:", file.path(boot_dir, "bootstrap_results.csv"), "\n")
````

````r
# ============================================================
# Bootstrap: Forest plot
# ============================================================
# Reuse display names from FIML section
bp_data <- boot_summary[boot_summary$label %in% names(display_names), ]
bp_data$display <- display_names[bp_data$label]
bp_data$display <- factor(bp_data$display, levels = rev(display_names[names(display_names) %in% bp_data$label]))

bp_data$effect_type <- ifelse(grepl("^a", bp_data$label), "a-path (SNP->M)",
                       ifelse(grepl("^b", bp_data$label), "b-path (M->Y)",
                       ifelse(bp_data$label == "cp", "Direct (c')",
                       ifelse(grepl("^ind", bp_data$label), "Specific indirect",
                       ifelse(bp_data$label == "total_indirect", "Total indirect",
                       ifelse(bp_data$label == "total", "Total effect", "Proportion"))))))

p_boot <- ggplot(bp_data, aes(x = boot_mean, y = display, color = effect_type)) +
    geom_vline(xintercept = 0, linetype = "dashed", color = "grey50") +
    geom_pointrange(aes(xmin = ci_lower, xmax = ci_upper), size = 0.5) +
    scale_color_brewer(palette = "Set1") +
    labs(title = "Bootstrap: Parallel Mediation Path Estimates",
         subtitle = paste0("N = ", N_total, " (FIML), ", B, " resamples (", n_converged, " converged)"),
         x = "Estimate (95% Percentile CI)", y = "", color = "Effect Type") +
    theme_minimal(base_size = 12) +
    theme(legend.position = "bottom")

ggsave(file.path(boot_dir, "bootstrap_forest_plot.png"), p_boot, width = 11, height = 9, dpi = 150)
ggsave(file.path(boot_dir, "bootstrap_forest_plot.pdf"), p_boot, width = 11, height = 9)
print(p_boot)
````

````r
# ============================================================
# Bootstrap: Distribution histograms for indirect effects
# ============================================================
ind_labels <- c(paste0("ind", 1:n_med), "total_indirect")
ind_display <- c(paste0("ind", 1:n_med, ": via ", MED_SHORT), "Total indirect")

plot_list <- list()
for (k in seq_along(ind_labels)) {
    lab <- ind_labels[k]
    vals <- boot_mat[, lab]
    vals <- vals[!is.na(vals)]
    if (length(vals) < 10) next
    
    df_hist <- data.frame(value = vals)
    p_h <- ggplot(df_hist, aes(x = value)) +
        geom_histogram(bins = 50, fill = "steelblue", alpha = 0.7) +
        geom_vline(xintercept = 0, linetype = "dashed", color = "red", size = 0.8) +
        geom_vline(xintercept = median(vals), linetype = "solid", color = "darkblue", size = 0.8) +
        labs(title = ind_display[k],
             x = "Indirect Effect", y = "Count") +
        theme_minimal(base_size = 10)
    plot_list[[k]] <- p_h
}

if (length(plot_list) > 0) {
    g <- do.call(gridExtra::grid.arrange, c(plot_list, ncol = 2))
    ggsave(file.path(boot_dir, "bootstrap_distributions_chr19_44938650_G_C.png"),
           g, width = 12, height = 10, dpi = 150)
    ggsave(file.path(boot_dir, "bootstrap_distributions_chr19_44938650_G_C.pdf"),
           g, width = 12, height = 10)
    cat("Bootstrap distribution plots saved.\n")
}
````

### Bootstrap Interpretation

The bootstrap provides **nonparametric percentile confidence intervals** for each path. 
Unlike FIML which assumes multivariate normality for the indirect effect CI, bootstrap CIs 
respect the actual sampling distribution of the product a*b, which is known to be non-normal.

Key comparison: If bootstrap CIs are consistent with FIML Wald CIs, this increases confidence 
in the FIML results. Discrepancies may indicate non-normality of the indirect effect distribution.


## 7. Method 3: MNAR Sensitivity Analysis

We assess robustness to Missing Not At Random (MNAR) assumptions by applying delta-adjustment:
shifting imputed values for missing mediators and outcome by multiples of their observed SD.

For the parallel model, we use a 2D grid: delta_mediators (applied uniformly to all 4 mediators) 
vs delta_outcome. At each grid point we impute missing values, fit the model with ML, 
and extract all indirect effects.


````r
# ============================================================
# MNAR Sensitivity: Delta-shift grid
# ============================================================
delta_range <- seq(-2, 2, length.out = 15)
grid <- expand.grid(delta_med = delta_range, delta_out = delta_range)
cat("MNAR grid:", nrow(grid), "points (", length(delta_range), "x", length(delta_range), ")\n")

# Pre-compute observed means and SDs
med_stats <- lapply(MED_COLS, function(m) {
    list(mean = mean(dat_sub[[m]], na.rm=TRUE), sd = sd(dat_sub[[m]], na.rm=TRUE))
})
names(med_stats) <- MED_COLS
out_mean <- mean(dat_sub[[OUT_COL]], na.rm=TRUE)
out_sd   <- sd(dat_sub[[OUT_COL]], na.rm=TRUE)
cov_means <- sapply(ALL_COVS, function(v) mean(dat_sub[[v]], na.rm=TRUE))

# Columns to store
mnar_labels <- c(paste0("ind", 1:n_med), "total_indirect", "cp", "total",
                 paste0("ind", 1:n_med, "_p"), "total_indirect_p")

mnar_results <- matrix(NA, nrow(grid), length(mnar_labels),
                       dimnames = list(NULL, mnar_labels))

cat("Running MNAR grid...\n")
t0 <- proc.time()[3]
n_ok <- 0

for (k in 1:nrow(grid)) {
    dat_imp <- dat_sub
    
    # Impute mediators
    for (m in MED_COLS) {
        na_idx <- is.na(dat_imp[[m]])
        dat_imp[[m]][na_idx] <- med_stats[[m]]$mean + grid$delta_med[k] * med_stats[[m]]$sd
    }
    
    # Impute outcome
    na_idx_y <- is.na(dat_imp[[OUT_COL]])
    dat_imp[[OUT_COL]][na_idx_y] <- out_mean + grid$delta_out[k] * out_sd
    
    # Impute covariates with column means
    for (cov in ALL_COVS) {
        dat_imp[[cov]][is.na(dat_imp[[cov]])] <- cov_means[cov]
    }
    
    fit_m <- tryCatch(
        sem(model_str, data = dat_imp, estimator = "ML", fixed.x = FALSE),
        error = function(e) NULL
    )
    
    if (!is.null(fit_m) && lavInspect(fit_m, "converged")) {
        pe_m <- parameterEstimates(fit_m)
        for (i in 1:n_med) {
            lab <- paste0("ind", i)
            row_m <- pe_m[pe_m$label == lab, ]
            if (nrow(row_m) == 1) {
                mnar_results[k, lab] <- row_m$est
                mnar_results[k, paste0(lab, "_p")] <- row_m$pvalue
            }
        }
        for (lab in c("total_indirect", "cp", "total")) {
            row_m <- pe_m[pe_m$label == lab, ]
            if (nrow(row_m) == 1) {
                mnar_results[k, lab] <- row_m$est
                if (lab == "total_indirect") mnar_results[k, "total_indirect_p"] <- row_m$pvalue
            }
        }
        n_ok <- n_ok + 1
    }
    
    if (k %% 50 == 0) {
        elapsed <- proc.time()[3] - t0
        cat(sprintf("  Grid point %d/%d (converged: %d, elapsed: %.0fs)\n", k, nrow(grid), n_ok, elapsed))
    }
}

cat(sprintf("\nMNAR complete: %d/%d grid points converged.\n", n_ok, nrow(grid)))

# Save full grid
mnar_df <- cbind(grid, as.data.frame(mnar_results))
mnar_df$N <- N_total
write.csv(mnar_df, file.path(mnar_dir, "mnar_grid_results.csv"), row.names = FALSE)
cat("Saved:", file.path(mnar_dir, "mnar_grid_results.csv"), "\n")
````

````r
# ============================================================
# MNAR: Tipping point analysis
# ============================================================
# For each indirect effect, find the minimum distance from (0,0) where it becomes non-significant

tipping <- data.frame(
    effect = character(),
    tipping_delta_med = numeric(),
    tipping_delta_out = numeric(),
    tipping_distance = numeric(),
    fiml_significant = logical(),
    stringsAsFactors = FALSE
)

for (i in 1:n_med) {
    lab_p <- paste0("ind", i, "_p")
    lab_e <- paste0("ind", i)
    
    # Check if FIML says it's significant at (0,0)
    ref_idx <- which(abs(grid$delta_med) < 0.01 & abs(grid$delta_out) < 0.01)
    fiml_sig <- FALSE
    if (length(ref_idx) > 0 && !is.na(mnar_results[ref_idx[1], lab_p])) {
        fiml_sig <- mnar_results[ref_idx[1], lab_p] < 0.05
    }
    
    # Find non-significant points
    ns_idx <- which(mnar_results[, lab_p] >= 0.05 & !is.na(mnar_results[, lab_p]))
    if (length(ns_idx) > 0 && fiml_sig) {
        dists <- sqrt(grid$delta_med[ns_idx]^2 + grid$delta_out[ns_idx]^2)
        min_idx <- ns_idx[which.min(dists)]
        tipping <- rbind(tipping, data.frame(
            effect = paste0("ind", i, " (", MED_SHORT[i], ")"),
            tipping_delta_med = grid$delta_med[min_idx],
            tipping_delta_out = grid$delta_out[min_idx],
            tipping_distance = min(dists),
            fiml_significant = TRUE,
            N = N_total
        ))
    } else {
        tipping <- rbind(tipping, data.frame(
            effect = paste0("ind", i, " (", MED_SHORT[i], ")"),
            tipping_delta_med = NA,
            tipping_delta_out = NA,
            tipping_distance = ifelse(fiml_sig, Inf, NA),
            fiml_significant = fiml_sig,
            N = N_total
        ))
    }
}

# Total indirect tipping
ref_idx <- which(abs(grid$delta_med) < 0.01 & abs(grid$delta_out) < 0.01)
ti_sig <- FALSE
if (length(ref_idx) > 0 && !is.na(mnar_results[ref_idx[1], "total_indirect_p"])) {
    ti_sig <- mnar_results[ref_idx[1], "total_indirect_p"] < 0.05
}
ns_idx_ti <- which(mnar_results[, "total_indirect_p"] >= 0.05 & !is.na(mnar_results[, "total_indirect_p"]))
if (length(ns_idx_ti) > 0 && ti_sig) {
    dists_ti <- sqrt(grid$delta_med[ns_idx_ti]^2 + grid$delta_out[ns_idx_ti]^2)
    min_idx_ti <- ns_idx_ti[which.min(dists_ti)]
    tipping <- rbind(tipping, data.frame(
        effect = "total_indirect",
        tipping_delta_med = grid$delta_med[min_idx_ti],
        tipping_delta_out = grid$delta_out[min_idx_ti],
        tipping_distance = min(dists_ti),
        fiml_significant = TRUE,
        N = N_total
    ))
} else {
    tipping <- rbind(tipping, data.frame(
        effect = "total_indirect",
        tipping_delta_med = NA, tipping_delta_out = NA,
        tipping_distance = ifelse(ti_sig, Inf, NA),
        fiml_significant = ti_sig,
        N = N_total
    ))
}

cat("\n=== Tipping Point Summary ===\n")
print(tipping, digits=3, row.names=FALSE)
write.csv(tipping, file.path(mnar_dir, "mnar_tipping_summary.csv"), row.names = FALSE)
````

### MNAR Interpretation

The MNAR sensitivity analysis tests what happens to our mediation results if missing data are 
**not missing at random**. The delta parameters represent how many SDs the missing values deviate 
from the observed mean:
- delta = 0: MAR assumption (same as FIML)
- delta > 0: missing subjects have higher values than observed
- delta < 0: missing subjects have lower values than observed

**Tipping point**: The minimum delta shift needed to make a significant result non-significant. 
Larger tipping distances indicate greater robustness.


## 8. Method 4: Bayesian SEM (blavaan/Stan)

Bayesian SEM provides posterior distributions for all parameters, enabling direct probability 
statements like P(indirect > 0). blavaan handles missing data via Stan data augmentation, 
using the full dataset.


````r
# ============================================================
# Bayesian SEM via blavaan
# ============================================================
suppressPackageStartupMessages({
    library(blavaan)
})

cat("Fitting Bayesian parallel mediation model...\n")
cat("N =", N_total, "(full data passed to blavaan with NAs)\n")
cat("Using 2 chains x 2000 samples (+ 1000 burnin) for efficiency with 4 mediators...\n\n")

fit_bayes <- tryCatch(
    bsem(model_str, data = dat_sub, fixed.x = FALSE, target = "stan",
         n.chains = 2, burnin = 1000, sample = 2000, seed = 42),
    error = function(e) { cat("Bayesian Error:", conditionMessage(e), "\n"); NULL }
)

if (!is.null(fit_bayes)) {
    cat("Bayesian model fit complete.\n\n")
    
    # Extract using parTable
    pt <- parTable(fit_bayes)
    labeled_pt <- pt[pt$label != "", c("label", "est", "se")]
    cat("=== Bayesian Parameter Estimates (parTable) ===\n")
    print(labeled_pt, digits=4, row.names=FALSE)
} else {
    cat("Bayesian model failed to fit.\n")
}
````

````r
# ============================================================
# Bayesian: Extract posterior draws and compute CIs
# ============================================================
if (!is.null(fit_bayes)) {
    draws <- tryCatch(
        do.call(rbind, blavInspect(fit_bayes, "mcmc")),
        error = function(e) {
            cat("Draw extraction via mcmc failed, trying draws...\n")
            tryCatch(do.call(rbind, blavInspect(fit_bayes, "draws")),
                     error = function(e2) NULL)
        }
    )
    
    if (!is.null(draws)) {
        cat("Posterior draws matrix:", nrow(draws), "x", ncol(draws), "\n")
        cat("Column names:", paste(head(colnames(draws), 20), collapse=", "), "...\n\n")
        
        # Map labels to draw columns
        # blavaan names draws by internal IDs; need to match via parTable
        pt_full <- parTable(fit_bayes)
        
        # For defined parameters (:=), compute from constituent draws
        # Get the regression parameters with labels
        labeled_rows <- pt_full[pt_full$label != "" & pt_full$op %in% c("~", "~~"), ]
        
        # Try direct column matching first
        bayes_results <- data.frame(
            label = character(), posterior_mean = numeric(), posterior_sd = numeric(),
            ci_lower = numeric(), ci_upper = numeric(), pp_direction = numeric(),
            stringsAsFactors = FALSE
        )
        
        # For defined (:=) parameters, we need to compute from draws
        # First get the raw path draws (a1, a2, a3, a4, b1, b2, b3, b4, cp)
        raw_labels <- c(paste0("a", 1:n_med), paste0("b", 1:n_med), "cp")
        raw_draws <- matrix(NA, nrow(draws), length(raw_labels), dimnames=list(NULL, raw_labels))
        
        for (lab in raw_labels) {
            # Find the parTable row
            pt_row <- pt_full[pt_full$label == lab & pt_full$op == "~", ]
            if (nrow(pt_row) > 0) {
                # The column in draws corresponds to the row index in parTable
                # blavaan uses "sig_i" or numbered columns
                pidx <- pt_row$plabel[1]
                # Try to find the column
                if (pidx %in% colnames(draws)) {
                    raw_draws[, lab] <- draws[, pidx]
                } else {
                    # Try by position - the free parameter index
                    free_idx <- pt_row$free[1]
                    if (free_idx > 0 && free_idx <= ncol(draws)) {
                        raw_draws[, lab] <- draws[, free_idx]
                    }
                }
            }
        }
        
        # Compute derived quantities
        derived_draws <- matrix(NA, nrow(draws), 0)
        derived_names <- c()
        
        # Specific indirects
        for (i in 1:n_med) {
            if (!any(is.na(raw_draws[, paste0("a", i)])) && !any(is.na(raw_draws[, paste0("b", i)]))) {
                ind_draws <- raw_draws[, paste0("a", i)] * raw_draws[, paste0("b", i)]
                derived_draws <- cbind(derived_draws, ind_draws)
                derived_names <- c(derived_names, paste0("ind", i))
            }
        }
        colnames(derived_draws) <- derived_names
        
        # Total indirect
        if (ncol(derived_draws) == n_med) {
            ti_draws <- rowSums(derived_draws[, paste0("ind", 1:n_med)])
            derived_draws <- cbind(derived_draws, total_indirect = ti_draws)
            derived_names <- c(derived_names, "total_indirect")
        }
        
        # Total
        if ("total_indirect" %in% colnames(derived_draws) && !any(is.na(raw_draws[, "cp"]))) {
            total_draws <- raw_draws[, "cp"] + derived_draws[, "total_indirect"]
            derived_draws <- cbind(derived_draws, total = total_draws)
            derived_names <- c(derived_names, "total")
        }
        
        # Prop mediated
        if ("total_indirect" %in% colnames(derived_draws) && "total" %in% colnames(derived_draws)) {
            pm_draws <- derived_draws[, "total_indirect"] / derived_draws[, "total"]
            # Clip extreme values
            pm_draws[abs(pm_draws) > 10] <- NA
            derived_draws <- cbind(derived_draws, prop_med = pm_draws)
            derived_names <- c(derived_names, "prop_med")
        }
        
        colnames(derived_draws) <- derived_names
        
        # Combine all draws
        all_draws <- cbind(raw_draws, derived_draws)
        
        # Build results table
        all_labels <- colnames(all_draws)
        for (lab in all_labels) {
            vals <- all_draws[, lab]
            vals <- vals[!is.na(vals)]
            if (length(vals) < 10) next
            
            bayes_results <- rbind(bayes_results, data.frame(
                label = lab,
                posterior_mean = mean(vals),
                posterior_sd = sd(vals),
                ci_lower = quantile(vals, 0.025),
                ci_upper = quantile(vals, 0.975),
                pp_direction = mean(vals > 0),
                stringsAsFactors = FALSE
            ))
        }
        rownames(bayes_results) <- NULL
        bayes_results$N <- N_total
        bayes_results$n_chains <- 2
        bayes_results$n_samples <- 2000
        
        cat("\n=== Bayesian Posterior Summaries ===\n")
        print(bayes_results, digits=4, row.names=FALSE)
        
        write.csv(bayes_results, file.path(bayes_dir, "bayesian_results.csv"), row.names = FALSE)
        cat("\nSaved:", file.path(bayes_dir, "bayesian_results.csv"), "\n")
    } else {
        cat("Could not extract posterior draws.\n")
    }
} else {
    cat("Skipping Bayesian results extraction (model did not fit).\n")
}
````

````r
# ============================================================
# Cross-method summary table
# ============================================================
main_labels <- c(paste0("a", 1:n_med), paste0("b", 1:n_med), "cp",
                 paste0("ind", 1:n_med), "total_indirect", "total", "prop_med")

summary_all <- data.frame()

# FIML
if (exists("key_pe")) {
    fiml_rows <- key_pe[key_pe$label %in% main_labels, ]
    if (nrow(fiml_rows) > 0) {
        summary_all <- rbind(summary_all, data.frame(
            exposure = SNP_COL,
            method = "FIML",
            label = fiml_rows$label,
            est = fiml_rows$est,
            se = fiml_rows$se,
            ci_lower = fiml_rows$ci.lower,
            ci_upper = fiml_rows$ci.upper,
            p_value = fiml_rows$pvalue,
            ci_type = "Wald",
            n_eff = paste0("N=", N_total, " (FIML)"),
            stringsAsFactors = FALSE
        ))
    }
}

# Bootstrap
if (exists("boot_summary")) {
    boot_rows <- boot_summary[boot_summary$label %in% main_labels, ]
    if (nrow(boot_rows) > 0) {
        summary_all <- rbind(summary_all, data.frame(
            exposure = SNP_COL,
            method = "Bootstrap",
            label = boot_rows$label,
            est = boot_rows$boot_mean,
            se = boot_rows$boot_se,
            ci_lower = boot_rows$ci_lower,
            ci_upper = boot_rows$ci_upper,
            p_value = boot_rows$p_value,
            ci_type = "Percentile",
            n_eff = paste0("N=", N_total, " (Bootstrap FIML, ", n_converged, "/", B, " converged)"),
            stringsAsFactors = FALSE
        ))
    }
}

# Bayesian
if (exists("bayes_results") && nrow(bayes_results) > 0) {
    bayes_rows <- bayes_results[bayes_results$label %in% main_labels, ]
    if (nrow(bayes_rows) > 0) {
        # For Bayesian, "p_value" = 2*min(PP, 1-PP) as a two-sided analog
        bayes_p <- 2 * pmin(bayes_rows$pp_direction, 1 - bayes_rows$pp_direction)
        summary_all <- rbind(summary_all, data.frame(
            exposure = SNP_COL,
            method = "Bayesian",
            label = bayes_rows$label,
            est = bayes_rows$posterior_mean,
            se = bayes_rows$posterior_sd,
            ci_lower = bayes_rows$ci_lower,
            ci_upper = bayes_rows$ci_upper,
            p_value = bayes_p,
            ci_type = "Credible",
            n_eff = paste0("N=", N_total, " (Bayesian, 2 chains x 2000)"),
            stringsAsFactors = FALSE
        ))
    }
}

cat("=== All Methods Summary ===\n")
print(summary_all, digits=4, row.names=FALSE)

write.csv(summary_all, file.path(summ_dir, "all_methods_summary.csv"), row.names = FALSE)
cat("\nSaved:", file.path(summ_dir, "all_methods_summary.csv"), "\n")
````

## 13. Post-Analysis Sanity Check

````r
# ============================================================
# Sanity checks
# ============================================================
cat("\n========================================\n")
cat("POST-ANALYSIS SANITY CHECK\n")
cat("========================================\n\n")

checks_passed <- c()
checks_warned <- c()
checks_failed <- c()

# 1. Sign consistency
cat("--- Check 1: Sign Consistency Across Methods ---\n")
for (lab in c(paste0("ind", 1:n_med), "total_indirect", "cp")) {
    signs <- sapply(c("FIML", "Bootstrap", "Bayesian"), function(m) {
        row <- summary_all[summary_all$method == m & summary_all$label == lab, ]
        if (nrow(row) == 1) return(sign(row$est)) else return(NA)
    })
    signs <- signs[!is.na(signs)]
    if (length(signs) > 1 && length(unique(signs[signs != 0])) > 1) {
        msg <- sprintf("SIGN DISAGREEMENT for %s: %s", lab, paste(names(signs), signs, sep="=", collapse=", "))
        cat(paste0("  [!] ", msg, "\n"))
        checks_failed <- c(checks_failed, msg)
    } else {
        cat(sprintf("  [ok] %s: all methods agree (sign=%s)\n", lab, 
                    ifelse(length(unique(signs)) == 1, ifelse(unique(signs) > 0, "+", "-"), "mixed/0")))
    }
}
if (length(checks_failed) == 0) checks_passed <- c(checks_passed, "Sign consistency: all methods agree")

# 2. Magnitude discrepancies
cat("\n--- Check 2: Magnitude Discrepancies ---\n")
mag_issues <- c()
for (lab in c(paste0("ind", 1:n_med), "total_indirect")) {
    ests <- sapply(c("FIML", "Bootstrap", "Bayesian"), function(m) {
        row <- summary_all[summary_all$method == m & summary_all$label == lab, ]
        if (nrow(row) == 1) return(abs(row$est)) else return(NA)
    })
    ests <- ests[!is.na(ests) & ests > 1e-8]
    if (length(ests) > 1 && max(ests) > 3 * min(ests)) {
        msg <- sprintf("Large magnitude discrepancy for %s (range: %.4f to %.4f)", lab, min(ests), max(ests))
        mag_issues <- c(mag_issues, msg)
        cat(paste0("  [!] ", msg, "\n"))
    }
}
if (length(mag_issues) == 0) {
    cat("  [ok] No large magnitude discrepancies detected.\n")
    checks_passed <- c(checks_passed, "Magnitude consistency: no >3x discrepancies")
}

# 3. Significance disagreements
cat("\n--- Check 3: Significance Disagreements ---\n")
for (lab in c(paste0("ind", 1:n_med), "total_indirect")) {
    sigs <- sapply(c("FIML", "Bootstrap", "Bayesian"), function(m) {
        row <- summary_all[summary_all$method == m & summary_all$label == lab, ]
        if (nrow(row) == 1 && !is.na(row$p_value)) return(row$p_value < 0.05) else return(NA)
    })
    sigs <- sigs[!is.na(sigs)]
    if (length(unique(sigs)) > 1) {
        msg <- sprintf("Significance disagreement for %s: %s", lab, 
                       paste(names(sigs), ifelse(sigs, "sig", "ns"), sep="=", collapse=", "))
        cat(paste0("  [warn] ", msg, "\n"))
        checks_warned <- c(checks_warned, msg)
    } else if (length(sigs) > 0) {
        cat(sprintf("  [ok] %s: all methods agree (%s)\n", lab, ifelse(sigs[1], "significant", "not significant")))
    }
}

# 4. Proportion mediated range
cat("\n--- Check 4: Proportion Mediated Range ---\n")
pm_rows <- summary_all[summary_all$label == "prop_med", ]
for (r in 1:nrow(pm_rows)) {
    pm_val <- pm_rows$est[r]
    meth <- pm_rows$method[r]
    if (!is.na(pm_val) && (pm_val < 0 || pm_val > 1)) {
        msg <- sprintf("Proportion mediated = %.3f (%s) -- outside [0,1], indicates suppression", pm_val, meth)
        cat(paste0("  [warn] ", msg, "\n"))
        checks_warned <- c(checks_warned, msg)
    } else if (!is.na(pm_val)) {
        cat(sprintf("  [ok] prop_med = %.3f (%s)\n", pm_val, meth))
    }
}

# 5. Large SEs
cat("\n--- Check 5: Very Large Standard Errors ---\n")
for (lab in c(paste0("ind", 1:n_med), "total_indirect")) {
    rows <- summary_all[summary_all$label == lab, ]
    for (r in 1:nrow(rows)) {
        if (!is.na(rows$se[r]) && !is.na(rows$est[r]) && abs(rows$est[r]) > 1e-8 &&
            rows$se[r] > 5 * abs(rows$est[r])) {
            msg <- sprintf("Large SE for %s (%s): SE=%.4f vs |est|=%.4f", lab, rows$method[r], rows$se[r], abs(rows$est[r]))
            cat(paste0("  [warn] ", msg, "\n"))
            checks_warned <- c(checks_warned, msg)
        }
    }
}
if (!any(grepl("Large SE", checks_warned))) {
    cat("  [ok] No excessively large standard errors.\n")
    checks_passed <- c(checks_passed, "Standard errors: no SE > 5x|estimate|")
}

# 7. Bootstrap failure rate
cat("\n--- Check 7: Bootstrap Convergence Rate ---\n")
fail_rate <- 1 - n_converged / B
if (fail_rate > 0.2) {
    msg <- sprintf("High bootstrap failure rate: %d/%d (%.1f%%) failed", B - n_converged, B, 100*fail_rate)
    checks_failed <- c(checks_failed, msg)
    cat(paste0("  [!] ", msg, "\n"))
} else {
    msg <- sprintf("Bootstrap convergence: %d/%d (%.1f%%)", n_converged, B, 100*n_converged/B)
    checks_passed <- c(checks_passed, msg)
    cat(paste0("  [ok] ", msg, "\n"))
}

# 8. MNAR tipping point
cat("\n--- Check 8: MNAR Tipping Point Distance ---\n")
if (exists("tipping") && nrow(tipping) > 0) {
    for (r in 1:nrow(tipping)) {
        if (!is.na(tipping$tipping_distance[r]) && tipping$tipping_distance[r] < 0.5) {
            msg <- sprintf("Close tipping point for %s: distance = %.2f SD", tipping$effect[r], tipping$tipping_distance[r])
            cat(paste0("  [warn] ", msg, "\n"))
            checks_warned <- c(checks_warned, msg)
        } else if (is.na(tipping$tipping_distance[r]) && tipping$fiml_significant[r]) {
            cat(sprintf("  [ok] %s: tipping point beyond grid (very robust)\n", tipping$effect[r]))
        } else if (!is.na(tipping$tipping_distance[r])) {
            cat(sprintf("  [ok] %s: tipping distance = %.2f SD\n", tipping$effect[r], tipping$tipping_distance[r]))
        }
    }
}

# 10. Suppression
cat("\n--- Check 10: Suppression Effects ---\n")
fiml_total <- summary_all[summary_all$method == "FIML" & summary_all$label == "total", "est"]
fiml_ti <- summary_all[summary_all$method == "FIML" & summary_all$label == "total_indirect", "est"]
fiml_cp <- summary_all[summary_all$method == "FIML" & summary_all$label == "cp", "est"]
if (length(fiml_ti) > 0 && length(fiml_cp) > 0 && !is.na(fiml_ti) && !is.na(fiml_cp)) {
    if (sign(fiml_ti) != sign(fiml_cp) && sign(fiml_ti) != 0 && sign(fiml_cp) != 0) {
        msg <- "Indirect and direct effects have OPPOSITE signs -- suppression pattern"
        cat(paste0("  [warn] ", msg, "\n"))
        checks_warned <- c(checks_warned, msg)
    } else {
        cat("  [ok] No suppression pattern detected.\n")
    }
}

# 11. Non-significant path with significant indirect
cat("\n--- Check 11: Component Path Check ---\n")
if (exists("key_pe")) {
    for (i in 1:n_med) {
        a_p <- key_pe[key_pe$label == paste0("a", i), "pvalue"]
        b_p <- key_pe[key_pe$label == paste0("b", i), "pvalue"]
        ind_p <- key_pe[key_pe$label == paste0("ind", i), "pvalue"]
        if (length(a_p) > 0 && length(b_p) > 0 && length(ind_p) > 0) {
            if (!is.na(ind_p) && ind_p < 0.05 && ((a_p > 0.1) || (b_p > 0.1))) {
                msg <- sprintf("ind%d significant (p=%.4f) but a%d (p=%.4f) or b%d (p=%.4f) non-significant", 
                              i, ind_p, i, a_p, i, b_p)
                cat(paste0("  [warn] ", msg, "\n"))
                checks_warned <- c(checks_warned, msg)
            }
        }
    }
}
cat("  [ok] Component paths checked.\n")

# Final summary
cat("\n========================================\n")
cat("SANITY CHECK SUMMARY\n")
cat("========================================\n")
cat(sprintf("Passed: %d\n", length(checks_passed)))
for (x in checks_passed) cat(paste0("  [ok] ", x, "\n"))
cat(sprintf("\nWarnings: %d\n", length(checks_warned)))
for (x in checks_warned) cat(paste0("  [warn] ", x, "\n"))
cat(sprintf("\nAnomalies: %d\n", length(checks_failed)))
for (x in checks_failed) cat(paste0("  [!] ", x, "\n"))
if (length(checks_failed) == 0 && length(checks_warned) == 0) {
    cat("\nAll sanity checks passed -- no anomalies detected.\n")
}
````
